```python
import zipfile
import os

# Create a destination folder
extract_to = "/content/unzipped_isro_data/"
os.makedirs(extract_to, exist_ok=True)

# Path to your folder in Drive
drive_path = "/content/drive/MyDrive/ISRO_Chandrayan3"

# Loop through and unzip all files in that folder
for file in os.listdir(drive_path):
    if file.endswith(".zip"):
        print(f"Unzipping {file}...")
        with zipfile.ZipFile(os.path.join(drive_path, file), 'r') as zip_ref:
            zip_ref.extractall(extract_to)

print("✅ All PDFs extracted to /content/unzipped_isro_data/")
```

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

# Use the LOCAL folder where we just unzipped everything
loader = PyPDFDirectoryLoader("/content/unzipped_isro_data/")
documents = loader.load()

# Re-run the splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=700, chunk_overlap=100)
chunks = text_splitter.split_documents(documents)

print(f"✅ FINAL SUCCESS: Loaded {len(documents)} pages and split into {len(chunks)} chunks.")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# 1. Initialize the Embedding Model (The 'Translator')
print("Converting 858 chunks into vectors... this will take a minute.")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 2. Create the FAISS Database
vector_store = FAISS.from_documents(chunks, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 3})

print("✅ SUCCESS: Your Chandrayaan-3 Search Engine is ready!")
```
